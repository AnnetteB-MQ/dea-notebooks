Air Quality Index conversion. Relevant to monitoring air hazard.

```python
data = [(0,0),
        (12,50),
        (35.4,100),
        (55.4,150),
        (150.4,200),
        (250.4,300),
        (350.4,400),
        (500.4,500)]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
top = 80

x = np.linspace(0, top, 200)
y = np.interp(x, [i for i,j in data], [j for i,j in data])
plt.plot(y, x)
plt.ylabel('PM2.5 µg/m³')
plt.xlabel('US AQI')

conc = plt.gca()
aqi = conc.twinx()
aqi.set_ylabel('Australia AQI')

aqi.axhline(200, c='r')
aqi.axhline(150, c='y')
aqi.axhline(100, c='y')
aqi.axhline(66, c='g')
aqi.axhline(33, c='g')

aqi.text(100, 200+5, 'Hazardous')
aqi.text(100, 150+5, 'Very Poor')
aqi.text(100, 100+5, 'Poor')
aqi.text(100, 66+5, 'Fair')
aqi.text(100, 33+5, 'Good')
aqi.text(100, 0+5, 'Very Good')

conc.set_ylim(0, top)
aqi.set_ylim(0, top*4)
plt.title('Smoky Air Quality Conversion')
```
