```python
#import the modules needed
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
#fix random seed
seed = 7
np.random.seed(seed)
```

```python
#load dataset with pands

dataset = pd.read_csv('/g/data/u46/users/ck9738/Datasets/ML_for_smad_green/training_datasets/AusAlps_training_set_3.csv', nrows=20000)
dataset
```

```python
features = dataset.as_matrix(columns=['t_blue', 't_green','t_red', 't_rededge1', 't_rededge2', 't_rededge3', 
                                                     't_nir1', 't_nir2', 't_swir1', 't_swir2',]).astype(float)
classes = dataset.as_matrix(columns=['class'])
```

```python
coded_classes = np_utils.to_categorical(classes)
features_train,features_test = np.split(features, 2)
classes_train,classes_test = np.split(coded_classes, 2)
answer = np.split(classes, 2)
```

```python

# define baseline model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim=10, activation='relu'))
    model.add(Dense(2, activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
#create the classififer using the defined baseline model
estimator = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=5, verbose=0)
```

```python
# set up kfold validation using 10 splits
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)

```

```python
estimator.fit(features_train,classes_train)
```

```python
results = estimator.predict(features_test)
```
