```python
import sys
import datacube
import numpy as np
import seaborn as sns
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
from datacube.utils import geometry
from datacube.utils.geometry import CRS
from datacube.helpers import write_geotiff
import os
```

```python
Snow_phen2016_2018=pd.read_csv('results/threshold_02/mask/phen/new/2016_2018_phenology.csv')
Snow_phen2014_2015=pd.read_csv('results/threshold_02/mask/phen/new/2014_2015_phenology.csv')
Snow_phen2012_2013=pd.read_csv('results/threshold_02/mask/phen/new/2012_2013_phenology.csv')
Snow_phen2010_2011=pd.read_csv('results/threshold_02/mask/phen/new/2010_2011_phenology.csv')

Snow_phen2010_2018=pd.concat([
    Snow_phen2010_2011, Snow_phen2012_2013, 
    Snow_phen2014_2015,
    Snow_phen2016_2018
])
```

```python
Snow_phen2008_2009=pd.read_csv('results/threshold_02/mask/phen/new/2008_2009_phenology.csv')
Snow_phen2006_2007=pd.read_csv('results/threshold_02/mask/phen/new/2006_2007_phenology.csv')
Snow_phen2004_2005=pd.read_csv('results/threshold_02/mask/phen/new/2004_2005_phenology.csv')
Snow_phen2002_2003=pd.read_csv('results/threshold_02/mask/phen/new/2002_2003_phenology.csv')
Snow_phen2000_2001=pd.read_csv('results/threshold_02/mask/phen/new/2000_2001_phenology.csv')

Snow_phen2000_2009=pd.concat([
    Snow_phen2000_2001, 
                              Snow_phen2002_2003, 
    Snow_phen2004_2005,Snow_phen2006_2007, 
    Snow_phen2008_2009
])
```

```python
Snow_phen1998_1999=pd.read_csv('results/threshold_02/mask/phen/new/1998_1999_phenology.csv')
Snow_phen1996_1997=pd.read_csv('results/threshold_02/mask/phen/new/1996_1997_phenology.csv')
Snow_phen1994_1995=pd.read_csv('results/threshold_02/mask/phen/new/1994_1995_phenology.csv')
Snow_phen1992_1993=pd.read_csv('results/threshold_02/mask/phen/new/1992_1993_phenology.csv')
Snow_phen1990_1991=pd.read_csv('results/threshold_02/mask/phen/new/1990_1991_phenology.csv')

Snow_phen1987_1989=pd.read_csv('results/threshold_02/mask/phen/new/1987_1989_phenology.csv')

Snow_phen1990_1999=pd.concat([
    Snow_phen1990_1991, 
    Snow_phen1992_1993, Snow_phen1994_1995,Snow_phen1996_1997,Snow_phen1998_1999])
```

```python
Snow_phen_all=pd.concat([Snow_phen1987_1989, 
    Snow_phen1990_1999, Snow_phen2000_2009,Snow_phen2010_2018])
Snow_phen_all.to_csv('phenology_1800_all.csv')
# snow_phen=[(Snow_phen2010_2018,2010),
#            (Snow_phen2000_2009, 2000), 
#            (Snow_phen1990_1999, 1990),
#            (Snow_phen1987_1989, 1980)
#           ]
```

```python
pdf = matplotlib.backends.backend_pdf.PdfPages("phenology_NSW_all.pdf")

fig, ax = plt.subplots(figsize=(14,5))
#     pdf = matplotlib.backends.backend_pdf.PdfPages("phenology_NSW.pdf")
sp = ax.scatter(Snow_phen_all['dayofyear'], #ndsi_ds.values, 
                Snow_phen_all['average_ndsi'], c=Snow_phen_all['year'],
                cmap='jet', marker='o', edgecolor='none',s=35)
plt.ylim(-1, 1)
plt.title('Snow Phenology above 1800 m (1987-2018)')
plt.xlabel('Days of the year')
plt.ylabel('ave NDSI for valid pixels over 1800m)')
fig.colorbar(sp)
pdf.savefig()
pdf.close()
```
